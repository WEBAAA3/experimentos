```
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn.svm import SVC
from sklearn.svm import LinearSVC
from sklearn import svm
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

# Carregar e Visualizar os dados do arquivo
Criei +1 coluna com valores binários a fim de otimizar o processamento.
Cachorro = 1 / Gato = 0

```
dados = pd.read_csv ("dogs_cats_bit.csv")
dados
```

## Exibindo os dados graficamente


```
sns.scatterplot(data= dados, x='comprimento', y ='peso', hue = 'classe')
```

```
x = dados[["comprimento", "peso"]]
y = dados["bit"]
```

Modelo **x = comprimento/peso**  vs **y = gato/cachorro**

```
# Fit the SVM model
modelo = svm.SVC(kernel='linear')
modelo.fit(x, y)
previsoes = modelo.predict(x)
```

Verificando a acurácia


```
accuracy_score(y, previsoes)
```

Traçando o hiperplano

```
# Hiperplano
w = modelo.coef_[0]
a = -w[0] / w[1]
xx = np.linspace(30, 60)
yy = a * xx - (modelo.intercept_[0]) / w[1]
```

```
# Plot the parallels to the separating hyperplane that pass through the support vectors
b = modelo.support_vectors_[0]
yy_down = a * xx + (b[1] - a * b[0])
b = modelo.support_vectors_[-1]
yy_up = a * xx + (b[1] - a * b[0])
```

Exibindo o gráfico

```

plt.plot(xx, yy, linewidth=2, color='black');
sns.scatterplot(data= dados, x='comprimento', y ='peso', hue = 'classe')
```
